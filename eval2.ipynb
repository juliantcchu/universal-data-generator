```python
from sdmetrics.reports.single_table import QualityReport
report = QualityReport()
```

```python
import pandas as pd

synthetic_data_multi = pd.read_csv('datasets/general_model_synthetic_bankChurn.csv')
synthetic_data_single = pd.read_csv('datasets/single_model_synthetic_bankChurn.csv')
synthetic_data_ctgan = pd.read_csv('datasets/ctgan_bankChurn.csv')

synthetic_data = synthetic_data_ctgan
# synthetic_data = pd.read_csv('datasets/synthetic_travelChurn_general_model.csv')

real_data = pd.read_csv('/Users/julianchu/Downloads/Churn Modeling.csv')
# real_data = pd.read_csv('/Users/julianchu/Downloads/Customertravel 3.csv')
# real_data = real_data.drop(['FrequentFlyer', 'AnnualIncomeClass', 'Target'], axis=1)


synthetic_data = synthetic_data[real_data.columns]

print(synthetic_data.dtypes)
print(real_data.dtypes)
```

```python
display(synthetic_data.head())
real_data.head()
```

```python
fields = {}
data_types = real_data.dtypes
for col in data_types.keys():
    if data_types[col] == 'object':
        fields[col] = {
            "type": "categorical",
        }
    elif data_types[col] == 'int64':
        fields[col] = {
            "type": "numerical",
            "subtype": "integer",
        }
    elif data_types[col] == 'float64':
        fields[col] = {
            "type": "numerical",
            "subtype": "float",
        }
metadata = {
    "primary_key": "RowNumber",
    "fields": fields
}

metadata
```

```python
report.generate(real_data, synthetic_data, metadata)
```

```python
fig = report.get_visualization(property_name='Column Shapes')
fig.show()
```
